# Task 1

 The NorthWind data set, provided in XML format on Moodle, includes details of orders received by a fictitious business and is
 based on a sample Microsoft SQL Server database.

     (a) Create a function to import this XML file. Your function must include all appropriate exception-handling clauses covering all possible error conditions. The function should return the parsed contents of the XML file.

```python
import xml.etree.ElementTree as ET

def Xmlfilecreation(file_path):
    try:
        tree = ET.parse(file_path)
        return tree
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except ET.ParseError:
        print(f"Error: The file '{file_path}' is not a well-formed XML.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
            
file_path = "NorthWind.xml"
xmlfile = Xmlfilecreation(file_path)
```

    (b) Using the function you created in a) above, load the XML file, assigning the parsed contents to a variable named parsed_xml. Then use the print() function to display the OrderID, CustomerID, ShipCity and ShipCountry for the 11th,14th, 17th and 20th orders. Hint: the range() function might come in useful here!


```python
parsed_xml = Xmlfilecreation("NorthWind.xml")
if parsed_xml is not None:
    root = parsed_xml.getroot()
    order_list = root.find('Orders').findall('Order')
    order_index = [10,13,16,19]
    for index in order_index:
        if index < len(order_list): 
            order = order_list[index]
            orderid = order.get('OrderID') 
            customerid = order.find('CustomerID').text
            shipcity = order.find('ShipCity').text
            shipcountry = order.find('ShipCountry').text
            print(f"OrderID: {orderid}, CustomerID: {customerid}, ShipCity: {shipcity}, ShipCountry: {shipcountry}")
        else:
            print(f"Order at index {index} does not exist.")
```

    (c).Using the parsed contents of the XML file in the parsed_xml variable created earlier, find all orders not taken by EmployeeIDs 5 or 6. Export these orders and their associated order details data for these orders to a CSV file. Your CSVfile should contain field names, using the tag names in the XML file. Note: you should not use Pandas for this task.

```python
import csv
if parsed_xml is not None:
    root = parsed_xml.getroot()
    order_list = root.find('Orders').findall('Order')
    csv_data = []
    for order in order_list:
        employe = order.find('EmployeeID').text
        if employe not in ['5', '6']:
            orderid = order.get('OrderID')
            customerid = order.find('CustomerID').text
            shipcity = order.find('ShipCity').text
            shipcountry = order.find('ShipCountry').text
            orderdetails = order.find('OrderDetails').findall('OrderDetail')
            for detail in orderdetails:
                productid = detail.find('ProductID').text
                unitprice = detail.find('UnitPrice').text
                quantity = detail.find('Quantity').text
                discount = detail.find('Discount').text
                csv_data.append([orderid, customerid, shipcity, shipcountry, productid, unitprice, quantity, discount])
    
    csv_file_name = 'filtered_orders.csv'
    with open(csv_file_name, mode='w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(['OrderID', 'CustomerID', 'ShipCity', 'ShipCountry', 'ProductID', 'UnitPrice', 'Quantity', 'Discount'])
        csv_writer.writerows(csv_data)
    
    print(f"Data exported to {csv_file_name} successfully.")
```

# Task 3

 Write a function that has one parameter that takes a string value as its argument. The function should use a regular expression
 to check if the string passed as an argument contains a private Class C IP address. The function should print a list of the
 private class C addresses found and should return the number of private Class C IP addresses found.
 Note: Private class C IP addresses are in the range 192.168.0.0 to 192.168.255.255.
 Test your function with multiple sample strings.

```python
import re

def private_class(samplstring):
    pattern = r'\b192\.168\.(\d{1,3})\.(\d{1,3})\b'
    matches = re.findall(pattern, samplstring)
    private_ips = [f"192.168.{match[0]}.{match[1]}" for match in matches]
    print(private_ips)
    return len(private_ips)

samplstring1 = "Text with some random IPs 192.168.1.1 and 192.168.200.255."
samplstring2 = "Another string with no private Class C only 172.16.0.1."
samplstring3 = "is these are IPs: 192.168.100.50 and 192.168.255.255."

# Call the function and print the results
print(f"Number of private Class C IPs: {private_class(samplstring1)}")
print(f"Number of private Class C IPs: {private_class(samplstring2)}")
print(f"Number of private Class C IPs: {private_class(samplstring3)}")
```
